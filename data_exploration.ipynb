```python
df_nan = df.isna().sum().sort_values(ascending=False)    
df_nan
```

Compute columns to drop

```python
drop_columns = df_nan.index[df_nan > 5042]
df.drop(drop_columns, axis=1, inplace=True)
df.shape
```

We drop a row if it contains 30% of values NaN

```python
df = df.dropna(thresh=0.7 * len(df.columns))
df.shape
```

```python
df.head(10)
```

Drop patient ID column and transform categorical variables into ordinal values

```python
df.drop(['Patient ID'], axis=1, inplace=True)
```

```python
def encode_categorical(df, categoricals): # TODO :- modify to ignore NaN
    """ Encodes the categorical featurs
    :param categorical_list: list of columns with not encoded categorical features
    :type categorical_list: list(string)
    """
    for category in categoricals:
        df[category] = df[category].astype('category')
        df[category] = df[category].cat.codes + 1
    return df
```

```python
categoricals = [
    'SARS-Cov-2 exam result',
    'Influenza A',
    'Influenza B',
    'Parainfluenza 1',
    'CoronavirusNL63',
    'Rhinovirus/Enterovirus',
    'Coronavirus HKU1',
    'Parainfluenza 3',
    'Chlamydophila pneumoniae',
    'Respiratory Syncytial Virus',
    'Adenovirus',
    'Parainfluenza 4',
    'Coronavirus229E',
    'CoronavirusOC43',
    'Inf A H1N1 2009',
    'Bordetella pertussis',
    'Metapneumovirus',
    'Parainfluenza 2',
    'Influenza B, rapid test',
    'Influenza A, rapid test',
]
```

```python
df = encode_categorical(df, categoricals)
df.head()
```

Fill missing value of each column with column average value 

```python
for column in df.columns:
    df[column].fillna(df[column].mean(), inplace=True)
```
